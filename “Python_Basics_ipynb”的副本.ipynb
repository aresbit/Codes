## **Python Basics**

## Math

```
a = 10 // 3            # → 3 (integer division)
b = 10 % 3             # → 1 (remainder of division)
c = 10 / 3             # → 3.3333333333333335 (automatic floating point)
d = (3 * 0.1) == 0.3   # → False (floats are not exact!)
e = 2 ** 1000          # (power function results in big number? no problem!)
print(a,b,c,d)
print(e)
1 + 1                  # → 2 (interactive math can use _ for last result)
```

## Strings & Lists

```
s = 'abcdefgh'
a = len(s)                      # → 8
b = s[0]                        # → 'a'
c = s[-1]                       # → 'h'
d = s[1:3]                      # → 'bc' (slicing)
e = s[3:-1]                     # → 'defg'
f = s[3:-1:2]                   # → 'df'
g = s[::-1]                     # → 'hgfedcba'
print(a,b,c,d,e,f,g)

x = list(range(5))              # → [0, 1, 2, 3, 4] (range = lazy evaluation)
y = [3, 5, 8]                   # → [3, 5, 8] (direct construction)
z = [i*i for i in range(1,6)]   # → [1, 4, 9, 16, 25] (list comprehension)
z.append(77)                    # (appending element to list)
z.extend([88,99])               # (extending list with another list)
print(x,y,z,z[3:6])       
```

## Sets & Dictionaries

```
s = set({2,1,3,2,1})                            # → {1, 2, 3}
s.add(4); s.remove(2)                           # → {1, 3, 4}
print(3 in s)
print(s)

d = {'alice': 24, 'bob': 22, 'charlie': 23}     # (dictionary)
d['eve'] = 26
v = d.pop('charlie',None)
print(d, d['alice'])
```

## Control



```
x = y = 2
if 2*x < y or x > 2*y:            # if, elif, else as usual
  print("far")
elif x == y:
  print("equal")
else:
  print("near")

for item in d:                    # traverse keys in dictionary (or list, set)              
  for i in range(len(d)):         # traverse over indexes
    for i, item in enumerate(d):  # both index and item
      break                       # break current loop

while x > 3:                      # while loops as usual
  print(x)
  x -= 1                          # x = x - 1
 
print("hello") if x == 5 else print("x =",x)  # conditional expression
```

## Functions

```
def square(x):                          
  return x*x

def next_few(x, number = 3):          # default of number is 3
  res = [] 
  for y in range(number):
    res.append(x+y)
  return res

print(next_few(3))                    # → [3, 4, 5] (x = 3, number = default)
first, *middle, last = next_few(3,5)  # first = 3, middle = [4, 5, 6], last = 7

def foo(var, *args, **d_args):        
  print(var)                          # var is just a variable of any type              
  print("args =", args)               # *args is an arbitrary list of variables
  print("key_args =", d_args)         # **d_args is a dictionary

foo(1, 2, 3, x=4, y=5)                # args = (2,3), d_args = {'x': 4, 'y': 5}
f = lambda x : x + 10                 # save space? lambda is one line function
print(f(5))                           # → 15
l = [i*i for i in range(1,6)]
l.sort(key=lambda v : v % 10)         # l = [1, 4, 25, 16, 9] (also max, etc.)
print(l)
```

## Other Types

```
b = True                    # boolean
bb = not b                  # bb = False
c = 2+3j                    # complex
cc = c-2                    # cc = 3j
t = (2,3)                   # tuple, like a list but immutable
d = {t: True, c: False}     # keys cannot be lists (tuples okay)
s = str(c)                  # s = '(2+3j)' (also int(), float(), etc.)
print(bb, cc, d, s)
```

## Conversion, Inspection

```
x = 'Hello'        # variable types can be changed whenever necessary
dir(x)             # attribute of a variable? even more info: help(x)
x = 10
print(x, type(x))  # type tells you the current type of x
x /= 3             # sometimes a variable changes its type automatically
print(x, type(x), isinstance(x,float))   # isinstance: check types/subtypes
x = int(x)         # you can also convert one type into another, if possible
print(x,type(x), isinstance(x,float))
```

## Variable Scope



```
x = y = z = 1       # globally: set all = 1

def foo():
  global x          # keyword global makes sure that the global x is used
  x, y = 2, 2       # x is global, y is local, z is unchanged (but accessible)
  print(x,y,z)      # → 2 2 1

foo()
print(x,y,z)        # → 2 1 1
```

## Shallow and Deep Copy

```
old = [1,[2,3]]
same = old
shallow = old.copy()         # copying basic elemements, referencing rest
import copy                  # deepcopy method must be imported
deep = copy.deepcopy(old)    # copying recursively
same[0] = 'a'
old[1][1] = 'c'
print('old =', old)          # → ['a', [2, 'c']]
print('same =', same)        # → ['a', [2, 'c']]
print('shallow =', shallow)  # → [1, [2, 'c']]
print('deep =', deep)        # → [1, [2, 3]]

def foo(v, l):               # beware about changing lists in functions
  v += 1
  l[1] = 'x'

x = 5
ll = [1,2]
foo(x,ll)
print("function:", x, ll)    # → 5 [1, 'x']
```

## Object Oriented Python

```
class Foo:
  def __init__(self, name):  # constructor
    self.name = name

  def printName(self):       # method (always self as first argument)
    print('I am', self.name)

bar = Foo('bar')             # new object (constructor needs name)
bar.printName()              # → I am bar
```
